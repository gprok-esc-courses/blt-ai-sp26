```python
import pandas as pd
from sklearn.cluster import KMeans, DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']] 	
```

```python
model = KMeans(n_clusters=3, random_state=42)
```

```python
df['cluster'] = model.fit_predict(X)
```

```python
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['cluster'])
```

```python
df['cluster'].unique()
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
model = KMeans(n_clusters=3, random_state=42)
```

```python
df['cluster_scaled'] = model.fit_predict(X_scaled)
```

```python
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['cluster_scaled'])
```

```python
df.head()
```

```python
model = DBSCAN(eps=0.6, min_samples=5)
```

```python
df['cluster_dbscan'] = model.fit_predict(X_scaled)
```

```python
plt.scatter(df['PetalLengthCm'], df['PetalWidthCm'], c=df['cluster_dbscan'])
```
